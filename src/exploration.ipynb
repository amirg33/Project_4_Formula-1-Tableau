```python
# Visualizing the data in order to see what kind of columns I have: 

import pandas as pd

# Dictionary of the CSV files
file_paths = {
    "circuits": "../data/Unprocessed/circuits.csv",
    "constructor_results": "../data/Unprocessed/constructor_results.csv",
    "constructor_standings": "../data/Unprocessed/constructor_standings.csv",
    "driver_standings": "../data/Unprocessed/driver_standings.csv",
    "drivers": "../data/Unprocessed/drivers.csv",
    "lap_times": "../data/Unprocessed/lap_times.csv",
    "pit_stops": "../data/Unprocessed/pit_stops.csv",
    "qualifying": "../data/Unprocessed/qualifying.csv",
    "races": "../data/Unprocessed/races.csv",
    "results": "../data/Unprocessed/results.csv",
    "seasons": "../data/Unprocessed/seasons.csv",
    "sprint_results": "../data/Unprocessed/sprint_results.csv",
    "constructors": "../data/Unprocessed/constructors.csv"
}
columns_df = pd.DataFrame()

# Iterate through the files and add their column names to the DataFrame
for file_name, file_path in file_paths.items():
    # Read the CSV file
    df = pd.read_csv(file_path)  
    # Add the columns to the DataFrame
    columns_df[file_name] = pd.Series(df.columns)  
```
